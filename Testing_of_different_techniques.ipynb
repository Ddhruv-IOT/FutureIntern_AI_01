```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load pre-trained model and tokenizer
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Encode input prompt
input_text = "The future of artificial intelligence is"
input_ids = tokenizer.encode(input_text, return_tensors="pt")

# Generate text using greedy search
greedy_output = model.generate(input_ids, max_length=50)

# Decode and print the generated text
output_text = tokenizer.decode(greedy_output[0], skip_special_tokens=True)
print("Greedy Search Output:\n", output_text)
```

```python
import matplotlib.pyplot as plt
from collections import Counter
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load pre-trained model and tokenizer
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Function to get n-gram frequencies
def get_ngram_frequencies(text, n=2):
    words = text.split()
    ngrams = zip(*[words[i:] for i in range(n)])
    return Counter(ngrams)

# Generate text without n-gram penalty
input_text = "The future of AI"
input_ids = tokenizer.encode(input_text, return_tensors="pt")
output_no_penalty = model.generate(input_ids, max_length=50)
text_no_penalty = tokenizer.decode(output_no_penalty[0], skip_special_tokens=True)

# Generate text with n-gram penalty
output_with_penalty = model.generate(input_ids, max_length=50, no_repeat_ngram_size=2)
text_with_penalty = tokenizer.decode(output_with_penalty[0], skip_special_tokens=True)

# Get n-gram frequencies
no_penalty_freqs = get_ngram_frequencies(text_no_penalty)
with_penalty_freqs = get_ngram_frequencies(text_with_penalty)

# Plotting the frequencies
labels, values_no_penalty = zip(*no_penalty_freqs.items())
values_with_penalty = [with_penalty_freqs.get(ngram, 0) for ngram in labels]

x = range(len(labels))

plt.figure(figsize=(12, 6))
plt.bar(x, values_no_penalty, width=0.4, label='No N-gram Penalty', align='center')
plt.bar(x, values_with_penalty, width=0.4, label='With N-gram Penalty', align='edge')
plt.xticks(x, [' '.join(label) for label in labels], rotation='vertical')
plt.xlabel('N-grams')
plt.ylabel('Frequency')
plt.title('N-gram Frequency Comparison')
plt.legend()
plt.show()
```

```python
# Load pre-trained model and tokenizer
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Encode input prompt
input_text = "The future of artificial intelligence is"
input_ids = tokenizer.encode(input_text, return_tensors="pt")

# Generate text using top-p sampling
top_p_output = model.generate(input_ids, max_length=50, do_sample=True, top_p=0.9)

# Decode and print the generated text
output_text = tokenizer.decode(top_p_output[0], skip_special_tokens=True)
print("Top-p Sampling Output:\n", output_text)
```

```python
import seaborn as sns
import numpy as np

# Generate text using different sampling strategies
output_top_k = model.generate(input_ids, max_length=50, do_sample=True, top_k=50)
output_top_p = model.generate(input_ids, max_length=50, do_sample=True, top_p=0.9)
output_temp = model.generate(input_ids, max_length=50, do_sample=True, temperature=0.7)

# Decode the outputs
text_top_k = tokenizer.decode(output_top_k[0], skip_special_tokens=True)
text_top_p = tokenizer.decode(output_top_p[0], skip_special_tokens=True)
text_temp = tokenizer.decode(output_temp[0], skip_special_tokens=True)

# Word lengths as a proxy for diversity
def word_length_distribution(text):
    words = text.split()
    return [len(word) for word in words]

top_k_lengths = word_length_distribution(text_top_k)
top_p_lengths = word_length_distribution(text_top_p)
temp_lengths = word_length_distribution(text_temp)

# Plotting word length distributions
plt.figure(figsize=(12, 6))
sns.kdeplot(top_k_lengths, label="Top-k Sampling", shade=True)
sns.kdeplot(top_p_lengths, label="Top-p Sampling", shade=True)
sns.kdeplot(temp_lengths, label="Temperature Sampling", shade=True)

plt.xlabel('Word Length')
plt.ylabel('Density')
plt.title('Word Length Distribution by Sampling Strategy')
plt.legend()
plt.show()
```
